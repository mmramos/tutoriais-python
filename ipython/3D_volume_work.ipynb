```python
nx = 10  #controla a dimensao x do volume (dimensao lateral)
ny = 10  #controla a dimensao y do volume (profundidade)
nz = 10  #controla a dimensao z do volume (numero de matrizes ou volume)

M = np.array([[[0]*nx]*ny]*nz)

#----------------------------------------------------------------------------#

M[0,:,:]=1 # Frente

M[nz-1,:,:]=2 #Tras

M[:,:,0]=3 # Esquerda

M[:,:,nx-1]=4 # Direita

M[:,0,:]=5 # Cima

M[:,ny-1,:]=6 # Baixo

#----------------------------------------------------------------------------#

log1 = [10,10,10,10,20,20,20,20,30,30]

log2 = [10,10,10,20,20,20,20,20,20,30]

log3 = [10,10,10,10,20,20,30,30,30,30]

#----------------------------------------------------------------------------#

M[2,:,2] = 7

M[4,:,6] = 7

M[7,:,4] = 7

#----------------------------------------------------------------------------#

M1 = np.array([[[0]*nx]*ny]*nz)

M1[2,:,2] = log1

M1[4,:,6] = log2

M1[7,:,4] = log3

M1[:,0,:]=10 # Cima

M1[:,ny-1,:]=30 # Baixo

FRE = np.array([log1]*nx).T

TRA = np.array([log3]*nx).T

DIR = np.array([log2]*nx)

M1[0,:,:]=FRE # Frente

M1[nz-1,:,:]=TRA #Tras


M1[:,:,0]=[log1]*nx # Esquerda

M1[:,:,nx-1]=DIR # Direita



M2 = M1

print M1

#----------------------------------------------------------------------------#
```

```python

for l in range(1000):
    for k in range(nz):
        for j in range(ny):
            for i in range(nx):
                
                if M[k,j,i] == 0:
                    M2[k,j,i] = (M1[k,j,i+1] + M1[k,j,i-1] + M1[k,j+1,i] + M1[k,j-1,i] + M1[k+1,j,i] + M1[k-1,j,i])/6
                else:
                    M2 = M1
                    
    M1 = M2
    
print M1
```

```python
#print M1

for k in range(nz):
    for j in range(ny):
        for i in range(nx):
            if M1[k,j,i] <= 14:
                M2[k,j,i] = 10
            if M1[k,j,i] >= 25:
                M2[k,j,i] = 30
            if M1[k,j,i] >= 15 and M1[k,j,i] <= 24:
                M2[k,j,i] = 20
                    
print M1
```
