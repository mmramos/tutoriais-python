```python
#--------------------------Importando os Pacotes-------------------------------#

%matplotlib inline
from pylab import *
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
```

```python
p = 100

t1 = []
t2 = []

for i in range (-p,p):
    t0 = i*0.001
    t1.append(t0)
    
    t02 = t0/3.0
    t2.append(t02)
    
t1 = np.array(t1,float)
t2 = np.array(t2,float)
    
print t1
print t2
```

```python
A = 1
F = 30.0

k = []

psi_o = []
psi_p = []

for i in range (p*2):
    
    EE = -(np.pi**2)*(F**2)*(t1[i]**2)
    psi_o0 = A*(1 - (2*(np.pi**2)) * (F**2) * (t1[i]**2))*np.exp(EE)
    psi_o.append(psi_o0)

    EE2 = -(np.pi**2)*(F**2)*(t2[i]**2)
    psi_p0 = A*(1 - (2*(np.pi**2)) * (F**2) * (t2[i]**2))*np.exp(EE2)
    psi_p.append(psi_p0)
    
    k0 = cos(t1[i]*100)
    k.append(k0)
    
#print EE
#print psi_o
```

```python
plt.plot(t1,psi_o)
plt.plot(t1,k,'-r')
plt.plot(t1,psi_p,'-g')
plt.show()
```

```python
ddd = np.convolve(psi_o,k,mode='same')

plt.plot(t1,ddd)
plt.show()
```
