```python
#--------------------------Importando os Pacotes-------------------------------#

%matplotlib inline
from pylab import *
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
```

```python
p = 100

t1 = np.linspace(0.00040,0.26000,p)
    
print t1
```

```python
A = 1
F = 30.0

td = 2*(1/F)

print td

k = []

psi_0 = []
psi_1 = []
psi_2 = []
psi_3 = []
psi_4 = []

for i in range (p):
    
    EE0 = -(np.pi**2)*(F**2)*((t1[i])**2)
    psi_00 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-0*td)**2))*np.exp(EE0)
    psi_0.append(psi_00)
    
for i in range (p):
    
    EE1 = -(np.pi**2)*(F**2)*((t1[i]-t1[50])**2)
    psi_01 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-t1[50])**2))*np.exp(EE1)
    psi_1.append(psi_01)
    
```

```python
#---------------------------------Limitando Intervalo-------------------------------#

psi_m = []
t_m = [] # Margem de tempo limite (maxima e minima)
i_m = [] # indices limites (maximo e minimo)

for i in range(len(psi_1)):
    if psi_1[i] <= -0.00001: # toda vez que psi_i for menor ou igual a 1x10^-5
        t_m.append(t1[i])
        i_m.append(i)
        
print i_m

i_m2 = np.linspace(min(i_m),max(i_m),max(i_m)-min(i_m)+1)
print i_m2
    
psi_m = [] #wavlet pura, sem bordas! AQUI!!!!
t_m2 = []

for i in range(len(i_m2)):
    psi_m.append(psi_1[35+i])
    t_m2.append(t1[35+i])
    
print psi_m
#print t_m2
```

```python
plt.plot(t1,psi_0)
plt.plot(t1,psi_1,'-g')
#plt.plot(t1,psi_2,'-b')
#plt.plot(t1,psi_3,'-g')
#plt.plot(t1,psi_4,'-b')
#plt.plot(t1,k,'-r')
plt.show()

plt.plot(t_m2,psi_m)
plt.show()
```

```python
J = p
I = p

psi_f = []
psi_r = [] # regressiva
psi_p = [] # progressiva

for j in range (J):
    for i in range (I):
        EE0 = -(np.pi**2)*(F**2)*((t1[i]-t1[j])**2)

        psi_00 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-t1[j])**2))*np.exp(EE0)
        psi_f.append(psi_00)
     
O = p
for j in range (O):
    for i in range (I):
        #correcao regressiva
        EER = -(np.pi**2)*(F**2)*((t1[i]+t1[j]+4.5*t1[0])**2)
        psi_0R = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]+t1[j]+4.5*t1[0])**2))*np.exp(EER) #nota: o termo "4.5*t1[0]" foi chutado
        
        psi_r.append(psi_0R)

O = p
for j in range (O):
    for i in range (I):
        #correcao progressiva
        EEP = -(np.pi**2)*(F**2)*((t1[i]-t1[j]-t1[p-1]-5.5*t1[0])**2)
        psi_0P = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-t1[j]-t1[p-1]-5.5*t1[0])**2))*np.exp(EEP) #nota: o termo "4.5*t1[0]" foi chutado
        
        psi_p.append(psi_0P)
        
psi_ff = np.reshape(psi_f, (J,I)) 
psi_s  = psi_ff.sum(axis=0)
print psi_s
      
##################################################################################
    
psi_fr = np.reshape(psi_r, (O,I)) # Vermelho - correcao da wavelet para tras
psi_rs = psi_fr.sum(axis=0)

psi_fp = np.reshape(psi_p, (O,I)) # Rosa - correcao da wavelet para frente
psi_ps = psi_fp.sum(axis=0)

###################################################################################

psi_srs = psi_s+psi_rs #Dados com correcao regressiva
psi_sps = psi_s+psi_ps #Dados com correcao progressiva
psi_sprs = psi_s+psi_rs+psi_ps

plt.plot(t1,psi_s)
plt.plot(t1,psi_rs,'-r')
plt.plot(t1,psi_ps,'-m')
plt.plot(t1,psi_srs,'-g')
plt.plot(t1,psi_sps,'-k')
plt.plot(t1,psi_sprs,'-y')

plt.show()
```
