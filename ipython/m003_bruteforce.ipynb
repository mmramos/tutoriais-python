```python
import numpy as np
```

```python
psh  = np.linspace(2.1,2.7,10)
dtsh = np.linspace(60,100,10)

pma = 2.66  # g/cm3 para o quartzo
pfl = 1.101 # g/cm3 para a água do mar

dtma = 72.0 # us/ft para o quartzo
dtfl = 62.5 # us/ft para a água do mar

ARGIL = 0.61
RHOb  = 2.29
DTP   = 75.5

print pma
print dtma
```

```python
jj = []
ii = []
ee = []
for i in range (len(dtsh)):
    for j in range (len(psh)):
        e = np.sqrt(((pma - RHOb)/(pma - pfl) - ARGIL*(pma - psh[j])/(pma - pfl) - (dtma - DTP)/(dtma - dtfl) - ARGIL*(dtma - dtsh[i])/(dtma - dtfl))**2)
        ii.append(i)
        jj.append(j)
        ee.append(e)


print ee
print '________________________________________________________________________________________________'

z = ee.index(min(ee))

print z,'posicao |',ee[z],'minimo ee |',dtsh[ii[z]],' dtsh |',psh[jj[z]],' psh'

```
