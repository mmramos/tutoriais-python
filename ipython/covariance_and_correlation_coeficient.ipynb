```python
import numpy as np
import matplotlib.pyplot as plt
import random
semente    = 10
random.seed(semente)
```

```python
x = np.linspace(1,50,50)
y = 3*x + 10
print(np.cov(x,y),'\n',)

plt.plot(x,y,'-b')
plt.plot(x,y,'ob')
plt.show()
```

```python
y2 = []
for i in range(len(x)):
    y2.append(3*x[i] + random.randint(-30,50))
print(np.cov(x,y2),'\n')

plt.plot(x,y,'-r')
plt.plot(x,y2,'or')
plt.show()
```
