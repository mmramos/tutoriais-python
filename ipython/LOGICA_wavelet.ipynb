```python
J = 100
I = 100

E = [] # termo com tudao

for j in range (J):
    for i in range (I):
        
        D = (2*i) - j # funcao de onda

        E.append(D)
    

F = np.reshape(E, (J,I))

print E,' -> E '
print F,' -> F '

S = F.sum(axis=0)
print S,' -> Soma'
```
