```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

#-------------------# Função 3D Linear #-------------------#

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

#-------------------# Função 3D Linear #-------------------#

ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

```python
mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

#-------------------# Função 3D Linear #-------------------#
a = 0
b = 0
c = 0

n = 5

r = n*n

x = []
for i in range (n):
    x0 = np.linspace(-2, 2, n)
    x.append(x0)

x = list (x)
x = np.array(x)
y = x.T



x = x.reshape(n*n,1)
y = y.reshape(n*n,1)

x = x.T
y = y.T

z = c + np.sqrt((r**2)-((x-a)**2)-((y-b)**2))

#-------------------# Função 3D Linear #-------------------#

ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

u = np.linspace(0, 2 * np.pi, n*n)
v = np.linspace(0, np.pi, n*n)

x = n * np.outer(np.cos(u), np.sin(v))
y = n * np.outer(np.sin(u), np.sin(v))
z = n * np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z,  rstride=4, cstride=4, color='b')

print x

plt.show()
```
