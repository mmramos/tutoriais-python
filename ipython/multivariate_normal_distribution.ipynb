Bivariate Distribution - for diagonal covariance
<br>Univariate $f_N(x) = \frac{1}{\sigma \sqrt{2 \pi}}exp[-\frac{1}{2}(x - \mu)^2/\sigma^2]$
<br>Bivariate  $f_N(x_1,x_2) = \frac{1}{\sigma_1 \sqrt{2 \pi}}exp[-\frac{1}{2}(x_1 - \mu_1)^2/\sigma_1 ^2] \times \frac{1}{\sigma_2 \sqrt{2 \pi}}exp[-\frac{1}{2}(x_2 - \mu_2)^2./\sigma_2 ^2]$
<br> Nota: Por ser uma covariância diagonal, os eixos batem. Mas caso não fosse, os eixos da distribuição estariam em ângulo

```python
def fn2(x1,x2,s1,s2,m1,m2):
    a = (-0.5*((x1 - m1)**2))/(s1**2)
    b = (-0.5*((x2 - m2)**2))/(s2**2)
    y = ((1/(s1*np.sqrt(2*np.pi)))*np.exp(a))*((1/(s2*np.sqrt(2*np.pi)))*np.exp(b))
    return y

xa = xb = np.linspace(0,20,21) 
print(xa)
sa = 2.5;sb = 3.15
ma = 7  ;mb = 10.5

# a = soma de 2 dados de 6 lados
# b = soma de 3 dados de 6 lados

[X1,X2] = np.meshgrid(xa,xb)
print(np.shape(X1))
M = np.zeros(np.shape(X1))

for j in range(21):
    for i in range(21):
        M[j,i] = fn2(X2[j,i],X1[j,i],sb,sa,mb,ma)

plt.imshow(M)
plt.colorbar()
plt.show()
print(np.max(M))
```
