```python
# - Equação que gera o mapa
import numpy as np

grid = [50,50]

CC = np.array([[0.0]*grid[0]]*grid[1],float)

for j in range (grid[1]):
    for i in range(grid[0]):
        CC[j,i] = 0.5859 + np.sqrt(np.cos(i/10.0000)**2 + np.cos(j/10.0000)**2)
```
