```python
#--------------------------Importando os Pacotes-------------------------------#

%matplotlib inline
from pylab import *
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
```

```python
p = 100

t1 = np.linspace(0.00040,0.26000,p)
    
print t1
```

```python
A = 1
F = 30.0

td = 2*(1/F)

print td

k = []

psi_0 = []
psi_1 = []
psi_2 = []
psi_3 = []
psi_4 = []

for i in range (p):
    
    EE0 = -(np.pi**2)*(F**2)*((t1[i]-0*td)**2)
    psi_00 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-0*td)**2))*np.exp(EE0)
    psi_0.append(psi_00)
    
for i in range (p): 
    EE1 = -(np.pi**2)*(F**2)*((t1[i]-1*td)**2)
    psi_01 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-1*td)**2))*np.exp(EE1)
    psi_1.append(psi_01)
    
for i in range (p): 
    EE2 = -(np.pi**2)*(F**2)*((t1[i]-2*td)**2)
    psi_02 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-2*td)**2))*np.exp(EE2)
    psi_2.append(psi_02)
    
for i in range (p): 
    EE3 = -(np.pi**2)*(F**2)*((t1[i]-3*td)**2)
    psi_03 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-3*td)**2))*np.exp(EE3)
    psi_3.append(psi_03)
    
for i in range (p): 
    EE4 = -(np.pi**2)*(F**2)*((t1[i]-4*td)**2)
    psi_04 = A*(1 - (2*(np.pi**2)) * (F**2) * ((t1[i]-4*td)**2))*np.exp(EE4)
    psi_4.append(psi_04)
    
    k0 = cos(t1[i]*100)
    k.append(k0)
```

```python
plt.plot(t1,psi_0)
plt.plot(t1,psi_1,'-g')
#plt.plot(t1,psi_2,'-b')
#plt.plot(t1,psi_3,'-g')
#plt.plot(t1,psi_4,'-b')
#plt.plot(t1,k,'-r')
plt.show()
```
