```python
import numpy as np
import matplotlib.pyplot as plt
import random
semente    = 10
random.seed(semente)
amostragem = 1000

def fn(x,s,m):
    a = (-0.5*((x - m)**2))/(s**2)
    y = (1/(s*np.sqrt(2*np.pi)))*np.exp(a)
    return y
```

```python
dado_1 = []

for i in range(amostragem):
    dado_1.append(random.randint(1,6))
    
plt.hist(dado_1,np.linspace(0,8,9))
plt.show()
```

$f_N(x) = \frac{1}{\sigma \sqrt{2 \pi}}exp[-\frac{1}{2}(x - \mu)^2/\sigma^2]$

```python
dado_2 = []

for i in range(amostragem):
    dado_2.append(random.randint(1,6)+random.randint(1,6))

hist02 = np.histogram(dado_2,np.linspace(2,12,11))

pd2 = [1/36,2/36,3/36,4/36,5/36,6/36,5/36,4/36,3/36,2/36]

hist2  = []
fdist2 = []
FN2    = []

for i in range(len(hist02[0])):
    hist2.append(hist02[1][i])
    fdist2.append(hist02[0][i]/sum(hist02[0])) # valor/total = distribui√ß√£o
    FN2.append(fn(hist02[1][i],2.5,7.0))
    
print ((hist02[0])) # Quantidades
print ((hist2))     # valores da soma dos 3 dados

print ('media = ',np.mean(hist2),'|desvio padr√£o = ',np.std(hist2))

plt.plot(hist2,fdist2,'ob') # dado observado
plt.plot(hist2,pd2,'-r')    # distribui√ß√£o padr√£o
plt.plot(hist2,FN2,'--g')   # Fun√ß√£o
plt.show()
```
