```python
%matplotlib inline
import numpy as np
import scipy as sp
from scipy import misc
import matplotlib.pyplot as plt
```

```python
m = misc.imread('2.png',mode='RGB')
g = m.shape
d1 = g[0]
d2 = g[1]

#print (m)
print (g)

m0 = m[:,:,0]
m1 = m[:,:,1]
m2 = m[:,:,2]
print (m2.shape)

plt.matshow(m0)
plt.colorbar()
plt.show()

plt.matshow(m1)
plt.colorbar()
plt.show()

plt.matshow(m2)
plt.colorbar()
plt.show()
```
